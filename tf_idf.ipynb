```python
import pandas as pd

df = pd.read_csv("books_data.csv")

df.head()
```

```python
df.isnull().sum()
```

```python
df["description"] = df["description"].fillna("")
df["categories"] = df["categories"].fillna("")
```

```python
df.isnull().sum()
```

```python
df["content"] = (
    df["description"] + " " +
    df["categories"]
)
```

```python
df["content"] = df["content"].str.lower()
```

```python
df.tail()
```

```python
import re
def remove_html(text):
    return re.sub(r'<.*?>','',str(text))
```

```python
df['content']=df['content'].apply(remove_html)
```

```python
def remove_url(text):
    return re.sub(r'https?://\S+|www\.\S+', '', text)

df["content"] = df["content"].apply(remove_url)
```

```python
import string

def remove_punctuation(text):
    return text.translate(
        str.maketrans('', '', string.punctuation)
    )

df["content"] = df["content"].apply(remove_punctuation)
```

```python
def remove_numbers(text):
    return re.sub(r'\d+', '', text)

df["content"] = df["content"].apply(remove_numbers)
```

```python
def remove_spaces(text):
    return re.sub(r'\s+', ' ', text).strip()

df["content"] = df["content"].apply(remove_spaces)
```

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
df["tokens"]=df["content"].apply(word_tokenize)
```

```python
from nltk.corpus import stopwords

nltk.download('stopw5ords')

stop_words = set(stopwords.words('english'))

df["tokens"] = df["tokens"].apply(
    lambda words: [
        word for word in words
        if word not in stop_words
    ]
)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

def lemmatize(tokens):

    doc = nlp(" ".join(tokens))

    return [
        token.lemma_
        for token in doc
    ]

df["tokens"] = df["tokens"].apply(lemmatize)
```

```python
df["processed_text"] = df["tokens"].apply(
    lambda x: " ".join(x)
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

tfidf_matrix = tfidf.fit_transform(
    df["processed_text"]
)
```

```python
import re
import string

def preprocess_query(text):

    text = str(text).lower()

    text = re.sub(r'<.*?>', '', text)

    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    text = text.translate(
        str.maketrans('', '', string.punctuation)
    )

    text = re.sub(r'\d+', '', text)

    text = re.sub(r'\s+', ' ', text).strip()

    return text
```
